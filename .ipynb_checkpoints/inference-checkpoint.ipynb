<h2>Model Inference</h2>

Run the cells and provide the required information in the menu.

Parameter Selection:

<b>Pipeline mode</b>:<br> 
Allow to do predictions given diferent traignings of the same model over the same dataset  or a single model prediction as usual.

<b>Weight Options (Multimodel) predictions</b>:<br><br> 
 <b>Last</b>:<br> Uses last.ckpt for each single model.<br>
 <b>Custom</b>:<br> Allow to choose independent .ckpt file for each model.<br>
 <b>Minimun</b>:<br> Look the .ckpt files with the minimun val_loss during training for each model.  

<b>YAML Path</b>:<br> 
Path to the inference configuration YAML file for prediction.<br><br>

<b>CKPT Path</b>:<br> Path to the model weights file corresponding to the trained model<br>

<b>Input base path</b>:<br>  Path to the input files. This should correspond to the out_path_base defined in the data wrangling notebook.

<span style="color:red;"><b>Note:</b></span><br>
If you want to the raw prediction of your model  leave the default predicted mode and desactive all the selected post processing options (Best option for model performance evaluation)

<b>Prediction mode</b>:<br> 
 Use the features of a <b>probabilistic model</b> taking n_samples from the distribution and apply the selected operation to generate a final prediction. The single option generate a regular prediction<br><br>
 <b> N Samples</b>:<br>
 Number of samples used acording with the prediction mode.<br>
This feature only has sense for take a multi prediction integration for a single image with a <b>probabilistic models</b> in other case n_samples identical samples will be use.<br><br>
  
 <b>Use Max projection</b>:<br> 
 If activated, the model will run inference on a maximum intensity projection of the $3D$ input volume. This projection collapses the $Z$ spatial dimension by taking the highest pixel value along that axis, creating a $(C,1,Y, X)$ image for the model to process.<br><br>

  <b>Apply Remove Small Objects</b>:<br>
  Removes small, isolated regions in the segmentation output. It requires one or more values representing the minimum volume in $\mu m^3$ an object must have to be kept. Objects smaller than this threshold will be deleted.<br>
  Enter a comma-separated list of sizes, one for each target object class. If you provide only a single number, that threshold will be applied to all object classes.<br><br>

  <b>Apply Small Holes Correction</b>:<br> 
  Fixes small errors by filling in holes within segmented objects. It requires one or more values representing the maximum area in $\mu m^2$ a hole must have to be filled. Holes larger than this threshold are considered genuine features and will remain open.<br>
  Enter a comma-separated list of areas, one for each target object class. If you provide only a single number, that threshold will be applied to all object classes.<br><br>

  <b>Apply Thickness Adjustment</b>:<br> 
  Applies topology-preserving thinning to the segmented objects. Its purpose is to reduce the thickness of the objects, to a minimal representation in $\mu m$ while ensuring that the connectivity and general shape of the original object are maintained.<br>
  Enter a comma-separated list of minimum thickness values, one for each target object class. If you provide only a single number, that threshold will be applied to all object classes.<br>
  To prevent the Thickness Adjustment (thinning) from being applied to a specific object class, set its corresponding minimum thickness value to a number significantly larger than the maximum possible thickness of any object in your volume (e.g., use 1000).<br> 
  This high value acts as a safeguard, ensuring the object's thickness constraint is never met, thus leaving the object unmodified.<br><br>

  <b>Apply Round Objetcs Correction (Noise prediction removal)</b>:<br>
  It is designed to remove small, round, and compact objects.<br><br>


 <span style="color:red;"><b>Shape Notes:</b></span>

 For the correct prediction the model require the input shape of the volumes in the format 
 $(C,Z,Y, X)$
